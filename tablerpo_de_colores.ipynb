```python
# imporamos las librerias que vamos a usar
import numpy as np
import matplotlib.pyplot as plt

# construimos la matriz que nos piden resolver#
A = np.array([
    [1/4, 0,   1/4, 2/4],
    [2/4, 1/4, 0,   1/4],
    [1/4, 2/4, 1/4, 0  ],
    [0,   1/4, 2/4, 1/4]
])

# construimos las condiciones inicial, vamos a suponer que va inciar todos en el sig estado
X0 = np.array([1, 0, 0, 0])

# vamos a realizar 20 iteraciones, se puede cambiar el numero de iteraciones si asi se quiere.
N = 20

#contruimos un alamacen para que guarden los resultados que es el siguiente
resultados = [X0]
X = X0
#construimos el ciclo for
for n in range(1, N+1):
    X = A @ X
    resultados.append(X)
    print(f"Iteración {n}: x={X[0]:.4f}, y={X[1]:.4f}, z={X[2]:.4f}, t={X[3]:.4f}") #  imprimimos las iteraciones cada una

# imprimimos el reusltado obtuvido
resultados = np.array(resultados)

#nos piden construir la grafica, ahora la vamos ha contruir
plt.figure(figsize=(10,6))
plt.plot(resultados[:,0], 'r-o', label='$x_n$')
plt.plot(resultados[:,1], 'g-o', label='$y_n$')
plt.plot(resultados[:,2], 'b-o', label='$z_n$')
plt.plot(resultados[:,3], 'm-o', label='$t_n$')

plt.title('Evolución de las probabilidades $x_n, y_n, z_n, t_n$')
plt.xlabel('Iteración n')
plt.ylabel('Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

#
#¿Son todos los sectores igual de probables cuando pase mucho tiempo?   #
#respuesta:en este caso si sucede, como se puede observar en las iteraciones de abajo,
# en la iteracion 14 ya todas los probabilidades son iguales#
```
